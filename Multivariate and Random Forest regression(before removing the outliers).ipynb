```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
y_train = df['UnitPrice']
y_train.dropna()
y_train
```

```python
features = ['InvoiceNo','StockCode','Description','Quantity','CustomerID']
x_train = pd.get_dummies(df[features])
x_train.dropna()
x_train
```

```python
td = pd.read_csv('Test.csv')
```

```python
x_test = pd.get_dummies(td[features])
x_test.dropna()
x_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
my_submission = pd.DataFrame({'UnitPrice': y_pred})
my_submission.to_csv('sub2.csv', index = False)
```

```python
from sklearn.ensemble import RandomForestRegressor
model1 = RandomForestRegressor(n_estimators = 100)
model1.fit(x_train, y_train)
```

```python
y_pred1 = model1.predict(x_test)
```

```python
my_submission = pd.DataFrame({'UnitPrice': y_pred1})
my_submission.to_csv('sub3.csv', index = False)
```
